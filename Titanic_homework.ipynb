Data Visualization


```python
plt.hist(data.Age,color ='Teal', edgecolor ='black')
plt.title('Age distribution of Titanic passengers')
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.axvline(data['Age'].mean(), linestyle ='dashdot',color='red')#averageline#
plt.axvline(data['Age'].min(), linestyle ='dashed',color='blue')
plt.axvline(data['Age'].max(), linestyle ='dashed',color='grey')
plt.grid(axis='x', linestyle='--', alpha=0.7)
```

```python

series_Pclass = data.Pclass.value_counts()
print(series_Pclass)
print(type(series_Pclass))
plt.pie(series_Pclass, labels =series_Pclass.index, autopct='%1.1f%%')
plt.title('Passengers of Titanic by class')
plt.show()
```
